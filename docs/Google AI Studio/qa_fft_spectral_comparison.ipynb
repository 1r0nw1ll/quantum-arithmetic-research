# QA FFT Spectral Comparison
Compare symbolic QA FFT to classic FFT on impulse input.

```
from sympy import Rational, cos, sin, pi, simplify

# QA Tuple: Only the central impulse is active
qa_tuple = (2, 1, 3, 4)

# Define theta_k for a few values of k
k_values = [1, 5, 11]
results = []

for k in k_values:
    theta_k = Rational(2 * pi * k, 24)
    cos_val = simplify(cos(theta_k).evalf())
    sin_val = simplify(sin(theta_k).evalf())
    H_k = simplify(qa_tuple[3] * cos(theta_k) + qa_tuple[2] * sin(theta_k))
    results.append((k, theta_k, cos_val, sin_val, H_k))

results
```

## Classic FFT Comparison
Using NumPy's FFT for a 5-sample impulse.

```
import numpy as np
import matplotlib.pyplot as plt

impulse = [0, 0, 1, 0, 0]
fft_result = np.fft.fft(impulse)

plt.stem(np.arange(5), np.abs(fft_result), use_line_collection=True)
plt.title("Classic FFT Magnitudes")
plt.xlabel("Frequency Bin")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```
